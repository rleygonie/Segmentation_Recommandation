```python
indexed_km = indexed.drop('CustomerID', 'Gender')
```

```python
assembler = VectorAssembler(inputCols = ["GenderIndexed","Age","Annual Income (k$)","Spending Score (1-100)"], outputCol = 'features')
```

```python
data = assembler.transform(indexed_km).select('features')
data.show(6)
```

```python
kmeans = KMeans().setK(5).setSeed(1)
kmeans.setMaxIter(300)
model = kmeans.fit(dataset)
model.setPredictionCol("Cluster")
predictions = model.transform(dataset)

evaluator = ClusteringEvaluator(predictionCol='Cluster')

silhouette = evaluator.evaluate(predictions)
print("Silhouette with squared euclidean distance = " + str(silhouette))
print("Distance Measure = ", model.getDistanceMeasure())
print("Max iterations = ", model.getMaxIter())

ctr=[]
centers = model.clusterCenters()
print("Cluster Centers: ")
for center in centers:
    ctr.append(center)
    print(center) 
```

```python
gender=[]
age=[]
salaire=[]
Spending_Score=[]
predi=[]
for i in range(len(pandasDF)):
    gender.append(pandasDF["features"][i][0])
    age.append(pandasDF["features"][i][1])
    salaire.append(pandasDF["features"][i][2])
    Spending_Score.append(pandasDF["features"][i][3])
    predi.append(pandasDF["Cluster"][i])
```

### Modèle Random Forest

```python
train, test = predictions.randomSplit([0.7, 0.3], seed = 2018)
```

```python
print('Shape train data:',(train.count(), len(train.columns)))
```

```python
print('Shape test data:',(test.count(), len(test.columns)))
```

```python
rf = RandomForestClassifier(labelCol="Cluster", featuresCol="features", numTrees=10)
```

```python
model_tree = rf.fit(train)
```

```python
predictions = model_tree.transform(test)
```

```python
predictions.show()
```

```python
evaluator = MulticlassClassificationEvaluator(labelCol="Cluster", predictionCol="prediction", metricName="accuracy")
accuracy = evaluator.evaluate(predictions)
print("Test Error = %g" % (1.0 - accuracy))
```
